```python
import pandas as pd
import glob
import os

path = r"D:\3A ENSAE\Machine Learning for climate risk\Project_Task\catnat_drought\catnat_drought"

files = glob.glob(os.path.join(path, "arrete_*.csv"))
```

```python
dfs = []

for f in files:
    df = pd.read_csv(f, sep=";")
    dfs.append(df)

arretes = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
arretes
```

```python
arretes_rec=arretes[arretes['Décision']=="Reconnue"]
arretes_rec
```

```python
arretes_rec.info()
```

```python
import pandas as pd

# Exemple pour Début d'Événement
arretes_rec["Début d'Événement"] = pd.to_datetime(
    arretes_rec["Début d'Événement"], dayfirst=True, errors="coerce"
)

# Optionnel : vérifier qu’il n’y a pas de NaT
print(arretes_rec["Début d'Événement"].isna().sum())


```

```python
arretes_rec["year"] = arretes_rec["Début d'Événement"].dt.year
```

```python
catnat_year = (
    arretes_rec
    .groupby(["N° Insee", "year"])
    .size()
    .reset_index(name="n_arretes")
)

catnat_year["catnat_drought"] = 1
```
