```python
import geopandas as gpd

data_err=gpd.read_file("D:\\3A ENSAE\\Machine Learning for climate risk\\Project\\Clean_Bases\\df_erreur.gpkg")

data_err
```

```python
data_err.columns
```

```python
cols_to_drop = [
    'swi_min', 'swi_mean', 'swi_max',
    'prenei_sum', 'preliq_sum', 'pe_sum', 'evap_sum',
    't_avg', 'ff_avg', 'q_avg', 'hu_avg',
    'dli_avg', 'ssi_avg', 'etp_avg',
    'tinf_avg', 'tsup_avg',
    'resr_neige_avg', 'resr_neige6_avg',
    'hteurneige_avg', 'hteurneige6_avg', 'hteurneigex_avg',
    'snow_frac_avg',
    'drainc_sum', 'runc_sum', 'ecoulement_sum',
    'wg_racine_avg', 'wgi_racine_avg','index__adm','geometry',
    'superficie','population'
]

data_err = data_err.drop(columns=cols_to_drop, errors='ignore')
data_err
```

```python
data_err.columns
```

```python
# index des lignes où le NIVEAU est maximal par commune-année
idx_niveau_max = (
    data_err
    .groupby(['code_insee', 'year'])['NIVEAU']
    .idxmax()
)

# agrégation principale
data_comm_annee = (
    data_err
    .groupby(['code_insee', 'year'], as_index=False)
    .agg(
        SWI_UNIF_MENS=('SWI_UNIF_MENS', 'min'),
        sq_error_rf=('sq_error_rf', 'max'),
        NIVEAU=('NIVEAU', 'max'),

        #cleabs=('cleabs', 'first'),
        #nom_officiel=('nom_officiel', 'first'),
        #nom_officiel_en_majuscules=('nom_officiel_en_majuscules', 'first'),
        nom_commune=('nom_commune','first'),
        code_insee_canton=('code_insee_canton', 'first'),
        code_insee_des_arrondissements=('code_insee_des_arrondissements', 'first'),
        code_insee_du_departement=('code_insee_du_departement', 'first'),
        code_insee_de_la_region=('code_insee_de_la_region', 'first'),
        #composition_du_canton=('composition_du_canton', 'first'),
        dept_name=('dept_name', 'first'),
        region_name=('region_name', 'first')
    )
)

# récupération de ALEA associée au NIVEAU max
alea_comm_annee = (
    data_err
    .loc[idx_niveau_max, ['code_insee', 'year', 'ALEA']]
)

# merge final
data_comm_annee = data_comm_annee.merge(
    alea_comm_annee,
    on=['code_insee', 'year'],
    how='left'
)
```

```python
import pandas as pd

catnat_final=pd.read_csv("D:\\3A ENSAE\\Machine Learning for climate risk\\Project\\Clean_Bases\\catnat_final.csv",sep=",")
catnat_final
```

```python
catnat_final['Insee'] = (
    catnat_final['Insee']
    .astype('Int64')     # gère les NA proprement
    .astype(str)
    .str.zfill(5)
)
```

```python
catnat_final=catnat_final[catnat_final['decision']=="Reconnue"]
catnat_final
```

```python
import pandas as pd

# Exemple pour Début d'Événement
catnat_final["debut_event"] = pd.to_datetime(
    catnat_final["fin_event"], dayfirst=True, errors="coerce"
)

# Optionnel : vérifier qu’il n’y a pas de NaT
print(catnat_final["debut_event"].isna().sum())


```

```python
catnat_final["year"] = catnat_final["debut_event"].dt.year
```

```python
catnat_final = (
    catnat_final
    .groupby(["Insee", "year"])
    .size()
    .reset_index(name="n_arretes")
)

catnat_final["catnat_drought"] = 1
```

```python
catnat_final['year'].value_counts()
```

```python
catnat_final.columns
```

```python
catnat_final = catnat_final.drop(columns=['n_arretes'], errors='ignore')
```

```python
df_merged = (
    data_comm_annee
    .merge(
        catnat_final[['Insee', 'year', 'catnat_drought']],
        left_on=['code_insee', 'year'],
        right_on=['Insee', 'year'],
        how='left'
    )
)

# Remplacer les NA par 0
df_merged['catnat_drought'] = df_merged['catnat_drought'].fillna(0).astype(int)

# Optionnel : supprimer la colonne clé en double
df_merged = df_merged.drop(columns=['Insee'])
```
